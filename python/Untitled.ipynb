```python
import re

path = "/home/serena/reuse/hom-generator/data/benchmarks/allTriangle/triangle-ga2.txt"
TIMESTAMP_PATTERN = "^TIME\s+(\d+)\|(.+)"


stamps = []
with open(path) as f:
    for l in f:
            match_obj = re.match(TIMESTAMP_PATTERN, l)
            if match_obj:
                stamps.append(match_obj.group(2))
                
muts = []

for s in stamps:
    if s.startswith("_mut"):
        muts.append(s.split()[0])
```

```python
print(len(muts))
muts
```

```python
seen = set()

for m in muts:
    lst = sorted(m.split(","))
    tup = tuple(lst)
    print(tup)
    if tup in seen:
        print(tup)
    else:
        seen.add(tup)
```
